```python
import matplotlib.patches as patches
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt

def draw_rectangle(ax, x, y, width, height, **kwargs):
    rectangle = patches.Rectangle((x, y), width, height, **kwargs)
    ax.add_patch(rectangle)
    return rectangle

def draw_line(ax, x, y, **kwargs):
    line = Line2D(x, y, **kwargs)
    ax.add_artist(line)

def draw_ellipse(ax, x, y, width, height, **kwargs):
    ellipse = patches.Ellipse((x, y), width, height, **kwargs)
    ax.add_patch(ellipse)

def draw_arc(ax, x, y, width, height, theta1, theta2, **kwargs):
    arc = patches.Arc((x, y), width, height, theta1=theta1, theta2=theta2, **kwargs)
    ax.add_patch(arc)
```

```python
def playing_surface():

    fig, ax = plt.subplots()
    
    #Did each dimensions by feet & inches. Therefore, all the numbers are based on feet.
    
    #The websites I used to find the dimensions were:
    #https://hockeymanitoba.ca/wp-content/uploads/2013/03/Rink-Marking-Diagrams.pdf
    #https://misschemchick.wordpress.com/2015/11/07/nhl-basics-the-playing-surface/
    #https://en.wikipedia.org/wiki/Ice_hockey_rink
    
    #To draw out the ice
    draw_rectangle(ax,x=28,y=0, width=144, height=85, color="white")
    draw_rectangle(ax,x=0,y=20.5, width=28, height=44, color="white") 
    draw_rectangle(ax,x=172,y=20.5, width=28, height=44, color="white")
    draw_ellipse(ax, x=28, y=20.5, width=56, height=41, color="white")
    draw_ellipse(ax, x=28, y=64.5, width=56, height=41, color="white")
    draw_ellipse(ax, x=172, y=20.5, width=56, height=41, color="white")
    draw_ellipse(ax, x=172, y=64.5, width=56, height=41, color="white")
    
    #To draw out the wall
    plt.plot([28, 172], [0, 0], color="brown")
    plt.plot([28, 172], [85, 85], color="brown")
    plt.plot([0, 0], [20.5, 64.5], color="brown")
    plt.plot([200, 200], [20.5, 64.5], color="brown")
    draw_arc(ax, x=28, y=20.5, width=56, height=41, theta1=180, theta2=270,color="brown")
    draw_arc(ax, x=28, y=64.5, width=56, height=41, theta1=90, theta2=180,color="brown")
    draw_arc(ax, x=172, y=20.5, width=56, height=41, theta1=270, theta2=360,color="brown")
    draw_arc(ax, x=172, y=64.5, width=56, height=41, theta1=0, theta2=90,color="brown")
    
    #To draw out the in-rink lines
    plt.plot([100, 100], [0, 85], color="red") #Center line
    plt.plot([75, 75], [0, 85], color="blue")#Left blue line
    plt.plot([125, 125], [0, 85], color="blue")#Right blue line
    
    #For the goal lines, I used Cal3 (MAST 218) to find the intersection points
    #between the elipses that I made from before to when x=11 & x=189. When 
    #x=11, the elipse intersections twice at y=4.21084 and at y= 80.78916.    
    plt.plot([11, 11], [4.21084, 80.78916], color="red") #Left goal line
    plt.plot([189, 189], [4.21084, 80.78916], color="red") #Right goal line
    
    #Center circle
    draw_arc(ax, x=100, y=42.5, width=30, height=30,theta1=0,theta2=360, color="blue")#Line
    draw_ellipse(ax, x=100, y=42.5, width=1, height=1, color="blue")#Dot
    
    #Left Red Small Circles
    draw_ellipse(ax, x=48, y=22, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=48, y=63, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=82, y=22, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=82, y=63, width=2, height=2, color="red")#Dot
    #Right Red Small Circles
    draw_ellipse(ax, x=118, y=22, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=118, y=63, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=152, y=22, width=2, height=2, color="red")#Dot
    draw_ellipse(ax, x=152, y=63, width=2, height=2, color="red")#Dot
    
    
    #Red circles
    draw_arc(ax, x=48, y=22, width=30, height=30,theta1=0,theta2=360, color="red")
    draw_arc(ax, x=48, y=63, width=30, height=30,theta1=0,theta2=360, color="red")
    draw_arc(ax, x=152, y=22, width=30, height=30,theta1=0,theta2=360, color="red")
    draw_arc(ax, x=152, y=63, width=30, height=30,theta1=0,theta2=360, color="red")
    
    #Goalie marks
    #Below, I'm drawing out the colours. My "weird" way of a semicircle
    draw_ellipse(ax, x=11, y=42.5, width=12, height=12, color="lightskyblue")
    draw_ellipse(ax, x=189, y=42.5, width=12, height=12, color="lightskyblue")
    draw_rectangle(ax,x=1,y=36.5, width=10, height=12, color="white")
    draw_rectangle(ax,x=189,y=36.5, width=10, height=12, color="white")
    draw_arc(ax, x=11, y=42.5, width=12, height=12,theta1=270,theta2=90, color="red") #Left goalie mark
    draw_arc(ax, x=189, y=42.5, width=12, height=12,theta1=90,theta2=270, color="red") #Left goalie mark
    
    #Scorekeeper semicircle
    draw_arc(ax, x=100, y=0, width=19.75, height=19.75,theta1=0,theta2=180, color="red")

    #Goalie trapezoid lines
    plt.plot([11, 0], [36.5, 31.5], color="red") #Bottom left line
    plt.plot([11, 0], [48.5, 53.5], color="red") #Top left  line
    plt.plot([189, 200], [36.5, 31.5], color="red") #Bottom right line
    plt.plot([189, 200], [48.5, 53.5], color="red") #Top left line

    ax.set_xlim([0, 200])
    ax.set_ylim([0, 85])
    plt.axis('off')
    plt.show()
    return fig, ax
```

```python
playing_surface()
```
